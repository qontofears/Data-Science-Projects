# Проект для «ТелеДом»

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

   <b>Описание услуг:</b><br>
Оператор предоставляет два основных типа услуг: <br>
Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.<br>
Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).<br>
    <b>Также доступны такие услуги:</b><br>
Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);<br>
Выделенная линия технической поддержки (TechSupport);<br>
Облачное хранилище файлов для резервного копирования данных (OnlineBackup);<br>
Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).<br>
Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

## Обучение моделей

```python
# Разделяем на выборки
features = data.drop('Active', axis=1)
target = data['Active']

features_train, features_test, target_train, target_test = train_test_split(features, target, 
                                                                            test_size=0.25, random_state=STATE)

numerics = list(features.select_dtypes(include=['int64', 'float64']).columns)
categorical = list(features.select_dtypes(include='object').columns)

transformer = ColumnTransformer(transformers=[
                            ('scaler', StandardScaler(), numerics),
                            ('ohe', OneHotEncoder(sparse=False, drop='first'), categorical)],remainder="passthrough")

transformer.fit(features_train)


features_train = transformer.transform(features_train)
features_test = transformer.transform(features_test)
```

```python
# Перебор гиперпараметров для LGBMClassifier
params = {
    'learning_rate': np.arange(0.01, 0.1, 0.01),
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'min_child_samples': [5, 10, 15],
 }

lgbm_model = LGBMClassifier(random_state=STATE)
gscvl = GridSearchCV(lgbm_model, params, scoring='roc_auc',cv=5, n_jobs=-1)
gscvl.fit(features_train, target_train)
```

```python
print(gscvl.best_score_) #метрика с кросс-валидации, которая нам и нужна
print(gscvl.best_params_) #лучшие гиперпараметры модели
```

Модель LGBM Classifier прошла порог, но обучалась дольше чем CatBoost, да и если подавать сырые признаки, то CatBoost справится лучше чем LGBM, поэтому эту модель мы отставляем.

## Выбор лучшей модели

```python
# Функция для графика метрики AUC-ROC
def AC (features, target, model):
    model = model
    probabilities = model.predict_proba(features)
    proba = probabilities[:, 1]

    fpr, tpr, thresholds = roc_curve(target, proba) 
    auc_roc = roc_auc_score(target, proba)

    print("Значение метрики:", auc_roc)
    plt.figure(figsize=(18, 10))
    plt.plot(fpr, tpr, linestyle='-')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.title('ROC-curve')
    plt.show()
    
```
